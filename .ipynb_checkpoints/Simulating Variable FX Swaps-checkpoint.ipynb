# Simulating FX Rollovers using Python

- A **rollover** in forex markets involves moving a position to the **following delivery date**, in which case the rollover **incurs a fee**.
- Depending on whether you have a long or short position, you may be due a **rollover credit** or **debit**.
- FX strategies that involve relatively **long hold periods** (weeks), can be impacted by the rollover (debit or credit) of the position.
- Due to the fact that the rollover of these positions **change over time**, it is important that this is **factored in** to our **backtests**, which allows for more **accurate simulations**.
- The **rollover rate** in forex is the **net interest return** on a currency position held **overnight** by a trader. 


## Background

- The roll entails the **cost of financing** an FX position. It is often derived from the central bank **interest differential** between the two currencies being traded, with an **additional fee from your broker**.
- Brokers usually apply the associated fees on a **daily basis**.
- Rollover can be **credited to or from** a trader's account.

## Simulating Historical FX Rollovers

The below graphs will show the rollover per standard FX Lot.

**1** FX Lot = **100,000 units** of the Quote Currency

```python
import pandas as pd
import matplotlib.pyplot as plt
import quandl
import warnings
warnings.filterwarnings("ignore")
```

For this example only **G8** counties will be used, although you may specify others.

```python
eur = quandl.get("BIS/PD_DXM")
usd = quandl.get("BIS/PD_DUS") # effective fed funds rate
gbp = quandl.get("BIS/PD_DGB")         
jpy = quandl.get("BIS/PD_DJP")
cad = quandl.get("BIS/PD_DCA")
chf = quandl.get("BIS/PD_DCH")
aud = quandl.get("BIS/PD_DAU")
nzd = quandl.get("BIS/PD_DNZ")
```

```python
def compute_rolls(base, quote, fee_broker):
    ird = 100000 * (base - quote) / (100*365) - fee_broker
    ird.columns = ["IRD"]
    
    ird.fillna(method="ffill", inplace=True)
    
    ird["roll_long"] = ird["IRD"] - fee_broker
    ird["roll_short"] = -ird["IRD"] - fee_broker
    
    return ird
```

## Historical Rollover of AUD/CAD 

```python
fee_broker = 5 # the amount a broker takes per lot of quote currency
aud_cad = compute_rolls(aud, cad, fee_broker)
aud_cad[["roll_long", "roll_short"]].dropna().plot(grid=True);
```

## Historical Rollover of AUD/NZD

```python
fee_broker = 5 
aud_nzd = compute_rolls(aud, nzd, fee_broker)
aud_nzd[["roll_long", "roll_short"]].dropna().plot(grid=True);
```

## Historical Rollover of EUR/USD

```python
fee_broker = 5 
eur_usd = compute_rolls(eur, usd, fee_broker)
eur_usd[["roll_long", "roll_short"]].dropna().plot(grid=True);
```

## Historical Rollover of EUR/JPY

```python
fee_broker = 5 
eur_jpy = compute_rolls(eur, jpy, fee_broker)
eur_jpy[["roll_long", "roll_short"]].dropna().plot(grid=True);
```

## Historical Rollover of CHF/JPY

```python
fee_broker = 5 
chf_jpy = compute_rolls(chf, jpy, fee_broker)
chf_jpy[["roll_long", "roll_short"]].dropna().plot(grid=True);
```

In summary, the above shows a simple way to estimate the historical **variable rollover costs** for FX positions.
